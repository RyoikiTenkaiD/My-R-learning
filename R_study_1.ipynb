```r
library(tidyverse)
glimpse(longley)

ggplot(longley, aes(x = Year, y = GNP)) + #assigns x and y from dataset
  geom_point(colour = "dark green") + #plots points and colours dark green
  geom_smooth(method = "loess", colour = "purple") + #creates non-linear regression line w/purple
  geom_smooth(method = "lm") + #creates linear regression line
  labs(y = "Gross National Product")
options(repr.plot.width = 12)
```
